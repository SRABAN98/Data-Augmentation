```python
import tensorflow as tf
```

```python
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import array_to_img,img_to_array,load_img

datagen = ImageDataGenerator(rotation_range = 40,
                            width_shift_range = 0.4,
                            height_shift_range = 0.4,
                            shear_range = 0.4,
                            zoom_range = 0.4,
                            horizontal_flip = True,
                            fill_mode = "nearest")
```

```python
img = load_img(r"C:\Users\dell\OneDrive\Documents\Data Science\8th\2.Data Augumention- DL\cat.jpg")
img
```

```python
x = img_to_array(img)
x = x.reshape((1,) + x.shape)

i = 0
for batch in datagen.flow(x,batch_size=1,
                          save_to_dir=(r"C:\Users\dell\OneDrive\Documents\Data Science\8th\2.Data Augumention- DL\DATA AUGUMENT"),save_prefix="cat",save_format="jpeg"):
    i+=1
    if i>10:
        break
```
